```python
#Reading the Original CSV

import pandas as pd
from pathlib import Path

import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine.url import URL
from config import CONNSTRING
engine=create_engine(CONNSTRING)

met_csv = Path("../MetObjects1.csv")
countries_csv = Path("../Countries.csv")

met_pd=pd.read_csv(met_csv)
countries_pd = pd.read_csv(countries_csv)

met_pd.head()
```

```python
met_pd.columns
```

```python
met_new_df= met_pd[[
 'Object ID',
 'Object End Date',
 'Title',
 'Department',
 'Culture',
 'Period',
 'Artist Display Name',
 'Artist Alpha Sort',
 'Artist Nationality',
 'Artist Gender',
 'Credit Line',
 'Geography Type',
 'City',
 'State',
 'County',
 'Country',
 'Region',
 'Subregion',
 'Classification',
 'Rights and Reproduction',
 'Link Resource',
 'Tags']]

met_new_df['Country_U'] = met_new_df['Country']

met_new_df.head()
```

```python
#Dropping the NaN countries
met_new_df = met_new_df.dropna(subset=['Country'])
met_new_df.head()
```

```python
# Uniques countries and format
unique_names = met_new_df['Country'].drop_duplicates()
print(unique_names)
```

```python
#country field in uppercase
met_new_df['Country_U'] = met_new_df['Country_U'].str.upper()
met_new_df.head()
```

```python
countries_pd.head()
```

```python
#Rename the columns
countries_pd.rename(columns={' name': 'name', ' nom': 'nom', ' iso2': 'iso2', ' iso3': 'iso3', ' phone_code': 'phone_code'}, inplace=True)
```

```python
#Adding a copy for country 
countries_pd['Country_U'] = countries_pd['name']
```

```python
#country field in uppercase
countries_pd['Country_U'] = countries_pd['Country_U'].str.upper()

countries_pd.head()
```

```python
# Uniques continets and format
unique_continents= countries_pd['continente'].drop_duplicates()
print(unique_continents)
```

```python
#Changing the continents to english for the map
new_continent = {'Europa': 'Europe', 'África': 'Africa', 'América': 'America', 'Antártida': 'Antarctica', 'Australia y Oceanía': 'Oceania'}

countries_pd['continente'] = countries_pd['continente'].replace(new_continent)

unique_continents= countries_pd['continente'].drop_duplicates()
print(unique_continents)
```

```python
countries_new_df= countries_pd[[
 'Country_U',
 'continente'
]]

countries_new_df.head()
```

```python
#Dropping country duplicates 
countries_new_df.drop_duplicates(inplace=True)
```

```python

# Replace the old country name with the new one
old_country_name = "ESTADOS FEDERADOS DE"
new_country_name = "MICRONESIA"
countries_new_df['Country_U'] = countries_new_df['Country_U'].apply(
    lambda x: new_country_name if x == old_country_name else x
)
```

```python
if 'UNITED STATES OF AMERICA' in countries_new_df['Country_U'].values:
    # Create a new DataFrame with the additional row
    new_row = pd.DataFrame([{'Country_U': 'UNITED STATES', 'continente': 'AMERICA'}])
    # Append the new row to countries_new_df
    countries_new_df = pd.concat([countries_new_df, new_row], ignore_index=True)
```

```python
countries_new_df.to_csv('clean_countries.csv', index=False)
```

```python
# Merge the datasets on the "Country_U" column
merged_met_new_df=pd.merge(met_new_df, countries_new_df[['Country_U', 'continente']], on='Country_U', how='left')
# Rename the "continente" column to "Continent"
merged_met_new_df.rename(columns={'continente': 'Continent'}, inplace=True)
# Save the updated dataset
merged_met_new_df.to_csv('met_clean.csv', sep='|', index=False)
```

```python

with engine.connect() as conn:
    merged_met_new_df.to_sql('met_clean',con=conn, index=False)
```
