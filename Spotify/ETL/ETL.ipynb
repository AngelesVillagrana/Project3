```python
#Reading the Original CSV

import pandas as pd
from pathlib import Path

import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine.url import URL
from config import CONNSTRING
engine= create_engine(CONNSTRING)

songs_pd = Path("../ETL/songs_normalize.csv")

songs_pd=pd.read_csv(songs_pd)


songs_pd.head()
```

```python
songs_pd.columns
```

```python
# Convert milliseconds to minutes
duration_minutes = songs_pd['duration_ms'] / 60000

# Insert the new column right after 'duration_ms' column
songs_pd.insert(songs_pd.columns.get_loc('duration_ms') + 1, 'duration_minutes', duration_minutes)

songs_pd.head()
```

```python
#Find the 10 songs that are the quietest of all
most_negative_loudness = songs_pd.nsmallest(10, 'loudness')
most_negative_loudness.head()
```

```python
highest_loudness = songs_pd.nlargest(10, 'loudness')
highest_loudness.head()
```

```python
highest_duration_minutes = songs_pd.nlargest(10, 'duration_minutes')
highest_duration_minutes.head()
```

```python
shortest_duration_minutes = songs_pd.nsmallest(10, 'duration_minutes')
shortest_duration_minutes.head()
```

```python
highest_danceability = songs_pd.nlargest(10, 'danceability')
highest_danceability.head()
```

```python
lowest_danceability = songs_pd.nsmallest(10, 'danceability')
lowest_danceability.head()
```

```python
highest_popularity = songs_pd.nlargest(10, 'popularity')
highest_popularity.head()
```

```python
lowest_popularity = songs_pd.nsmallest(10, 'popularity')
lowest_popularity.head()
```

```python

with engine.connect() as conn:
    songs_pd.to_sql('songs',con=conn, index=False)
```

```python
songs_pd.to_csv('songs.csv', sep='|', index=False)

```
