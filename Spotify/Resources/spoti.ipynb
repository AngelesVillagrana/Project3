```python
import pandas as pd
from pathlib import Path
import webbrowser


from IPython.display import display, Javascript




#bokeh
from bokeh.io import output_notebook, show, output_file
from bokeh.models import ColumnDataSource, HoverTool, TapTool, CustomJS
from bokeh.plotting import figure
from bokeh.layouts import column
from bokeh.models.widgets import DataTable, TableColumn
from bokeh.models.callbacks import CustomJS


spoti_csv = Path("/Users/juliocc/Desktop/project-3/songs_normalize.csv")
spoti_pd=pd.read_csv(spoti_csv)

spoti_pd.head(10)
```

```python
#enlistar generos unicos

unique_genres_count = spoti_pd['genre'].nunique()
print(f"Number of unique genres: {unique_genres_count}")

# List the unique values in the 'genre' column
unique_genres = spoti_pd['genre'].unique()
#print(f"Unique genres: {unique_genres}")
```

```python
#ARTISTAS POR CANTIDAD DE APARICIONES
artist_counts = spoti_pd['artist'].value_counts()

artist_counts_df = artist_counts.reset_index()
artist_counts_df.columns = ['artist', 'count']

artist_counts_df = artist_counts_df.sort_values(by='count', ascending=False)

# Display the result
print(artist_counts_df)
```

```python
#dropeamos 1998 y 2020 que tienen vlores incompletos y acotamos medicion a 20 años
spoti_df = spoti_pd[(spoti_pd['year'] != 1998) & (spoti_pd['year'] != 2020)]
spoti_df = spoti_df.reset_index(drop=True)
```

```python

#PARA CREAR UNA LINEA DE TIEMPO QUE TE MUESTRE  EL TOP10  DE CANCIONES POR AÑO
#diccionario con el top 10 x año

top_10_songs_by_year = {}
for year in spoti_df['year'].unique():
    yearly_data = spoti_df[spoti_df['year'] == year]
    top_10_songs = yearly_data.nlargest(10, 'popularity')[['song', 'artist', 'popularity']]
    top_10_songs_by_year[year] = top_10_songs

for year, songs in top_10_songs_by_year.items():
    print(f"Year: {year}")
    print(songs)
```

```python
output_notebook()

# Preparamos data
yearly_counts = spoti_df['year'].value_counts().sort_index()

# Create a ColumnDataSource for the timeline
source = ColumnDataSource(data=dict(year=yearly_counts.index, count=yearly_counts.values))

# Create a ColumnDataSource for the DataTable (initially empty)
song_source = ColumnDataSource(data=dict(song=[], artist=[], popularity=[]))

# Create the timeline figure
p = figure(title='Spotify Top Songs Timeline (2000-2019)', x_axis_label='Year', y_axis_label='Number of Songs',
           tools="tap", height=400, width=800)
p.vbar(x='year', top='count', width=0.9, source=source)

# Hover que muestre año y cantidad x año)
hover = HoverTool(tooltips=[('Year', '@year'), ('Count', '@count')])
p.add_tools(hover)

# DataTable con campos para mostrar el top10 (quitar popularity?*)

columns = [TableColumn(field='song', title='Song'),
           TableColumn(field='artist', title='Artist')]
data_table = DataTable(source=song_source, columns=columns, width=800)

# JavaScript callback to update the DataTable
callback = CustomJS(args=dict(source=source, song_source=song_source, top_10_songsr=top_10_songs), code="""
    var indices = source.selected.indices;
    if (indices.length > 0) {
        var year = source.data['year'][indices[0]];
        var songs = top_10_songs[year];

        song_source.data = {
            song: songs['song'],
            artist: songs['artist'],
            popularity: songs['popularity']
        };
        song_source.change.emit();
    }
""")


taptool = p.select(type=TapTool)
taptool.callback = callback


#solo una barra seleccionable
source.selected.js_on_change('indices', CustomJS(args=dict(source=source), code="""
    var indices = source.selected.indices;
    if (indices.length > 1) {
        source.selected.indices = [indices[indices.length - 1]];
    }
"""))


# Layout the plot and the table
layout = column(p, data_table)

# Show the plot
output_file("spotify_top_songs_timeline.html")
show(layout)


# Open the HTML file in the default web browser
webbrowser.open("spotify_top_songs_timeline.html")
```
